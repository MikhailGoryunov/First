```python
import numpy as np
import pandas as pd
import datetime
from datetime import datetime, date
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 80)

table = pd.read_excel('T:\Ручки\Py\export1.xlsx',
                        index_col='Материал')
sel = table.Завод == 2500
#(len(sel), sel.sum())
table = table[~sel]
sel1 = table.Завод ==2501
#(len(sel1), sel1.sum())
table = table[~sel1]
sel2 = table.Завод ==2900
table = table[~sel2]

#table

reindexed = table.reset_index()
t = reindexed.set_index(['Завод', 'Материал'])
t.rename(columns=
        {'Кол-во в ЕИ ввода': 'Количество'},
        inplace=True)
t.rename(columns=
        {'Дата проводки': 'Дата'},
        inplace=True)
copy = t.copy()
t = copy.drop(['Склад', 'Вид движения', 'Особый запас', 'Документ материала', 'Поз. док. материала', 'ЕИ ввода'], axis = 1)
#t
#grouped_by_Завод = at.groupby('Завод')
#grouped_by_Завод
#grouped_by_Материал = at.groupby('Материал')
#grouped_by_Материал

#def print_groups (group_object):
#    for name, group in group_object:
#        print(name)
#        print(group[:5])
#print_groups(grouped_by_Завод)
#print_groups(grouped_by_Материал)

tab = t.groupby(level=['Завод', 'Материал'])
#print_groups(at.groupby(level=['Завод', 'Материал']))

t = tab.agg({'Количество': np.sum,
        'Дата' : np.max})

salony = pd.read_excel("T:\Ручки\Образцы\Таблица образцов в салонах клиентов.xlsx",
                sheet_name='Для программы',
                #usecols=['Артикул', 'Филиал'],
                  index_col='Артикул')
salony = salony.reset_index()
salony.rename(columns=
        {'Филиал': 'Завод'},
        inplace=True)
salony.rename(columns=
        {'Артикул': 'Материал'},
        inplace=True)
reindexed = salony.reset_index()
salony = reindexed.set_index(['Завод', 'Материал'])
#salony = salony[['Кол-во в салонах']]
salony.rename(columns=
        {'Количество': 'В салонах'},
        inplace=True)
#salony = salony.groupby(level=['Завод', 'Материал'])
salony = salony[['В салонах']]
#def print_groups (group_object):
#    for name, group in group_object:
#        print (name)
#        print(group[:10])
#print_groups(salony)
#salony = salony.agg({'В салонах': np.sum})
#salony = salony.reset_index()
#salony
#t[:10]
tab1 = t.join(salony, how='outer')
#tab1 = tab1[['Склад', 'Дата проводки', 'Количество']]
tab1['В салонах'] = tab1['В салонах'].fillna(0)
tab1 = tab1[tab1['В салонах'] == 0]
t1 = tab1[tab1.Количество > 0]

t1['time'] = pd.Timestamp('now')

#t1['time'] = pd.Timestamp('00:00')#datetime.now().time() 

#t
#datetime.now().date()
#tab1_copy = tab1.copy()
#tab1_copy['time'] = pd.Timestamp('now') #datetime.now().date()
#t_copy
#pd.to_datetime(['2019-01-21'], errors="coerce")
#t_copy.dtypes
#t1['Дата проводки'] = t1.to_datetime(t1['Дата проводки'])
#t1['time'] = t1.to_datetime(t1['Дата проводки'])
#t1
t1['Время'] = t1.time - t1.Дата
t1 = t1[t1.Время > '14 days']
t1.rename(columns=
        {'Дата': 'Дата последнего прихода'},
        inplace=True)
t1.rename(columns=
        {'Время': 'Лежит, дней'},
        inplace=True)
t1 = t1[['Количество', 'Дата последнего прихода', 'Лежит, дней']]
t1
```

```python
u = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Union Knopf',
                index_col='Article')
e = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Eureka',
                index_col='Article')
g = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Giusti',
                index_col='Article')
m = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Marella design',
                index_col='Article')
b = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Boyard',
                index_col='Article')
v = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Валмакс',
                index_col='Article')
c = pd.read_excel("T:\разное\Действительно нужные документы\Прайс-лист РУЧКИ с 01 01 2019.xlsx",
                sheet_name='Cosma',
                index_col='Article')
res = pd.concat([u, e, g, m, b, v, c], sort=True)
res.rename(columns=
        {'Новая цена для Фиеры Трейдинг': 'Цена ФТ'},
        inplace=True)
res.index.names = ['Материал']
res = res[['Себестоимость', 'Цена ФТ']]
res = res.round(2)
res
```
